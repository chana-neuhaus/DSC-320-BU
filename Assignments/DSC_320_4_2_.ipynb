###Average Rate of Change

```python
def rate_of_change(f, a, b):
  return (f(b) - f(a))/ (b - a)
```

### Average Rate of Change to Instantaneous Rate of Change

```python
def s(t):
  return 4.9 * (t ** 2)

avg_speed = rate_of_change(s, 5, 6)
print(f"Average speed between 5 and 6 Seconds: {avg_speed}")
```

```python
avg_speed_two = rate_of_change(s, 5, 5.5)
print(f"Average speed between 5 and 5.5 seconds: {avg_speed_two}")
```

```python
avg_speed_three = rate_of_change(s, 5, 5.1)
print(f"Average speed between 5 and 5.1 seconds: {avg_speed_three}")
```

```python
inst_speed = rate_of_change(s, 5, 5.00001)
print(f"Approximate Instantaneous Speed: {inst_speed}")
```

###Calculating and Interpreting Partial Derivatives

```python
def p(c, y):
  return 16000 + (2400 * c) - (1800 * y)
```

```python
print(f"Selling price at 5 years, rating 8: ${p(8,5)}")
```

∂C/∂P is 2400. This means that with every rating point increase, the predicted selling price increases by $2400, assuming age stays constant.
	​


∂P/∂Y is - 1800. This means that with every added year, the predicted selling price of the car decreases by $1800, assuming condition stays constatn.
