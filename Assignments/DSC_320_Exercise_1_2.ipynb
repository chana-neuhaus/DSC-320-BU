```python
#importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#defining RMSE function
def rmse(y_true, y_preds):
  '''Takes an array of true values and an array of predicted values and
  returns Root Mean Squared Error'''

  #count of values in array:
  n = y_true.size

  #calculates squared residuals by squaring the difference of the predicted values and true values
  sum_res_squared = np.sum((y_preds - y_true)**2)
  #calculating Root Mean Squared Error by taking square root of the sum of residuals divided by n
  rmse = np.sqrt(sum_res_squared/n)
  return rmse
```

```python
#defining MAE function
def mae(y_true,y_preds):
  '''Takes array of true values and array of predicted values and
  returns Mean Absolute Error'''

  #count of values in array:
  n = y_true.size

  #calculates sum of absolute value of residuals
  sum_abs_res = np.sum(np.abs(y_preds-y_true))

  #calculates Mean Absolute Error by dividing the sum of the absolute residuals by n

  mae = sum_abs_res/n
  return(mae)
```

```python
#Defining accuracy function
def accuracy(actual,predicted):
  '''Takes array of actual values and predicted (Boolean) values
  and returns accuracy'''

  #creates an array of Boolean values where actual is same as predicted
  bool_array = actual == predicted

  #count of accurate predictions
  accurate = np.sum(bool_array)

  #calculates accuracy by dividing count of accuate predictions by the size of the Boolean array
  accuracy = accurate/bool_array.size
  return accuracy
```

```python
#Importing csv dataset as Pandas dataframe
df = pd.read_csv("housing_data.csv")

#displaying dataframe
df
```

```python
#converting sale price and predicted sale price columns to nupy arrays
y_true = df["sale_price"].to_numpy()
y_pred = df["sale_price_pred"].to_numpy()
print(y_true.shape)
print(y_pred.shape)
```

```python
#Finding RMSE of data
rmse(y_true, y_pred)
```

```python
#Finding MAE of data
mae(y_true, y_pred)
```
