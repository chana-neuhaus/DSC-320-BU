```python
#importing dependencies
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#defining Riemann sum function
#takes a function, a and b of interval, and n (for number of rectangles) as arguments
def R(f,a,b,n):
  dx = (b-a)/n
  total = 0
  for i in range(n):
    i += 1
    y = f(a + (i * dx))
    total += y * dx
  return total
```

```python
#defining function f(x)
def f(x):
  return math.sqrt(4 - (x ** 2))
```

```python
#building values for rectangles

#calculating delta x (for width of rectangle)
dx = (2 - 0)/4

#creating numpy array with right endpoints (based on dx)
right_endpoints = np.array([0.5, 1.0, 1.5, 2.0])
#evaluating function at each right endpoint to find heights of rectangle
heights = [f(x) for x in right_endpoints]

print(f"dx = {dx}")
print(f"right endpoints: {right_endpoints}")
print(f"heights: {heights}")
```

```python
#plotting curve
plt.plot(x, y)

#plotting rectangles basd on endpoints, heights and widths
plt.bar(
    right_endpoints - dx,
    heights,
    width = dx,
    align = "edge",
    alpha = 0.3,
    edgecolor = "black"
)
plt.xlim(0,2.1)
plt.ylim(0, 2.1)
plt.show()
```

If we look at the last height value, it is 0. There is therefore no rectangle visible at the last right endpoint. If we calculate the area of each of these rectangles (by multiplying the width by the height), and add them up, we get a Riemann sum of approximately 2.5.

Alternatively, we can use the Riemann sum formula: (function defined above)

```python
#applying Riemann sum function
R(f,0,2,4)
```

```python
#evaluating Riemann sum at increasingly higher intervals
print(R(f,0,2,4))
print(R(f,0,2,10))
print(R(f,0,2,100))
print(R(f,0,2,1000))
print(R(f,0,2,10000))
print(R(f,0,2,100000))
print(R(f,0,2,1000000))
```

We observe the Riemann sum converging to approximately 3.14.

This value is also known as pi, and is the ratio of a circle's circumference to its diameter. As we noticed from our previous graph, our function f(x) represents the upper-right quarter of a  circle (where the radius would be 2). The area of the the curve should be equal to one quarter of the area of the circle, and so we would expect the Riemann sum to converge to pi.
