```python
#importing dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

##1. Data Normalization

```python
#creating normalizer function
def normalize(vector):
  normalized_vector = (vector - np.min(vector))/(np.max(vector)-np.min(vector))
  return normalized_vector
```

##2. Data Standardization

```python
#creating standardization function
def standardize(vector):
  mean = np.average(vector)
  stdev = np.std(vector)
  standardized_vector = (vector - mean)/stdev
  return standardized_vector
```
